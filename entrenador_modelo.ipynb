```python
#Importar la data
import pandas as pd

df = pd.read_csv('data_2021_2024.csv')
```

```python
# Se selcción solo algunos parametros importantes del documento de la fiscalía
data_c = df[['fecha_hecho', 'hora_hecho', 'categoria_delito', 'alcaldia_hecho']]
```

```python
# Se concatenan la fecha y la hora del hecho'
data_c['fecha_hecho'] = data_c.fecha_hecho.astype(str).str.cat(data_c.hora_hecho.astype(str), sep=',')
data_c = data_c.drop(['hora_hecho'], axis=1)
# Convertimos la unión de la fecha y la hora en datetime64
data_c['fecha_hecho'] = pd.to_datetime(data_c['fecha_hecho'], format='%d/%m/%y,%H:%M:%S', errors='coerce')
```

```python
from pycaret.classification import *

# Preparamos el setup para poder entrenar el modelo
modelo = setup(data_c, target='categoria_delito', categorical_features=['alcaldia_hecho'],
          date_features=['fecha_hecho'], create_date_columns=["year", "month", "day", "hour"], )
```

```python
# Se comparan diferentes modelos de machine learning
best_c = compare_models(sort='F1')
```
